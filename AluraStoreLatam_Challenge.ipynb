### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
lista_tiendas = [tienda1, tienda2, tienda3, tienda4]
tiendas_ventas_total = []
cont_tienda = 0

for tienda in lista_tiendas:
  cont_tienda += 1
  tiendas_ventas_total.append(tienda.Precio.sum())
  print(f'La facturación total de la tienda {cont_tienda}  es de {tienda.Precio.sum()}')
```

```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
plt.bar(tiendas, tiendas_ventas_total)
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.title('Ventas por tienda')
plt.show()
```

```python
print(f'La tienda 1 es la que posee mas facturación con un total de {tienda1.Precio.sum()}')
print(f'La tienda 4 es la que posee menos facturación con un total de {tienda4.Precio.sum()}')
```

# 3. Calificación promedio de la tienda


```python
calificacion_general =  tiendas_combinadas['Calificación'].mean()
print(f'la calificacion general de las tiendas es de {round(calificacion_general,2)}%')
```

```python
calificaciones_tiendas = [tienda['Calificación'].mean() for tienda in lista_tiendas]

for i in range (len(calificaciones_tiendas)):
  print(f'La calificacion de satisfacción de la tienda {i+1} es de {round(calificaciones_tiendas[i],3)}%')
```

La tienda con menos satisfacción es la tienda 1 y siguiente de este, la tienda 4.



# 5. Envío promedio por tienda

```python
promedios_tiendas = [tienda['Costo de envío'].mean() for tienda in lista_tiendas]

for i in range (len(promedios_tiendas)):
  print(f'El costo de envío promedio de la tienda {i+1} es de {round(promedios_tiendas[i],2)}')
```
